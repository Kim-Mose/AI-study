# SVM (Support Vector Machine)
1. **정의** : 데이터를 가장 잘 분리하는 결정 경계(초평면)를 찾는 알고리즘

2. **주요 용어**

| 용어 | 설명 |
| --- | --- |
| 초평면 (Hyperplane) | 데이터를 두 클래스로 나누는 경계 |
| 서포트 벡터 | 결정 경계와 가장 가까운 데이터 |
| 마진 (Margin) | 결정 경계와 서포트 벡터 사이의 거리 |
| 커널 (Kernel) | 데이터를 고차원으로 변환하는 함수 |

3. **핵심 아이디어**
    - 마진이 가장 큰 결정 경계를 찾는다.
    - 마진이 클수록 새로운 데이터에 대한 일반화 성능이 좋아진다.

## 하드 마진 vs 소프트 마진
- **하드 마진** : 모든 데이터를 완벽히 분리 (이상치에 민감, 선형 분리 불가능하면 사용 불가)
- **소프트 마진** : 약간의 오차를 허용 (이상치에 강함, 실무에서 주로 사용)

소프트 마진의 정도는 하이퍼파라미터 $C$로 조절한다.
- $C$가 크다 : 오차를 적게 허용 (하드 마진에 가까움) → 과적합 위험
- $C$가 작다 : 오차를 많이 허용 (소프트 마진) → 과소적합 위험

## 커널 트릭 (Kernel Trick)
선형으로 분리되지 않는 데이터를 고차원으로 보내면 선형 분리가 가능해진다.

이때 실제로 고차원으로 변환하지 않고 내적 결과만 계산하는 방법을 커널 트릭이라고 한다.

| 커널 | 설명 |
| --- | --- |
| linear | 선형 결정 경계 |
| poly | 다항식 커널 |
| rbf | 가우시안 커널 (기본값, 가장 많이 사용) |
| sigmoid | 시그모이드 커널 |

## Sklearn으로 SVM 구현해보기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()

X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)

model = SVC()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(accuracy_score(y_test, y_pred))
```

## SVM의 하이퍼파라미터 알아보기

| 파라미터명 | 기본값 | 설명 |
| --- | --- | --- |
| C | 1.0 | 오차 허용 정도 (클수록 하드 마진에 가까움) |
| kernel | "rbf" | 커널 종류 (linear, poly, rbf, sigmoid) |
| gamma | "scale" | rbf, poly, sigmoid 커널의 영향 범위 |
| degree | 3 | poly 커널의 차수 |
| random_state | None | 랜덤 시드 |
