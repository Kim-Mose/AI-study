# 머신러닝 개요 (Machine Learning Overview)
머신러닝이란 데이터를 기반으로 패턴을 학습하고, 학습한 패턴을 바탕으로 예측하는 기술이다.

## AI vs ML vs DL
- **인공지능(AI)**: 사람의 지능을 모방하는 모든 기술
- **머신러닝(ML)**: AI의 한 분야로, 데이터를 통해 학습하는 방법
- **딥러닝(DL)**: ML의 한 분야로, 인공 신경망을 사용하는 방법

$$AI \supset ML \supset DL$$

# 지도학습 (Supervised Learning)
정답(Label)이 있는 데이터로 학습하는 방법이다.
입력(X)과 정답(y)의 관계를 학습하여 새로운 입력에 대해 정답을 예측한다.

## 회귀 (Regression)
연속적인 숫자를 예측하는 문제
- 집값 예측, 온도 예측, 매출 예측 등
- 대표 알고리즘: 선형 회귀, 다항 회귀, Ridge, Lasso

## 분류 (Classification)
데이터가 어떤 카테고리에 속하는지 예측하는 문제
- 스팸 메일 분류, 질병 진단, 이미지 분류 등
- 대표 알고리즘: 로지스틱 회귀, SVM, KNN, 의사결정나무, 랜덤포레스트

| 구분 | 회귀 | 분류 |
| --- | --- | --- |
| 출력 | 연속적인 숫자 | 카테고리(클래스) |
| 예시 | 집값: 35000만원 | 합격/불합격 |
| 평가 지표 | MSE, RMSE, R² | Accuracy, Precision, Recall, F1 |

# 비지도학습 (Unsupervised Learning)
정답(Label)이 없는 데이터에서 패턴이나 구조를 스스로 찾아내는 방법이다.

## 군집화 (Clustering)
비슷한 데이터끼리 그룹으로 묶는 것
- 고객 세분화, 문서 분류, 이상 탐지 등
- 대표 알고리즘: K-Means, DBSCAN, 계층적 군집화

## 차원 축소 (Dimensionality Reduction)
데이터의 특성(feature)이 너무 많을 때 중요한 특성만 남기는 것
- 데이터 시각화, 노이즈 제거, 학습 속도 개선 등
- 대표 알고리즘: PCA, t-SNE

| 구분 | 군집화 | 차원 축소 |
| --- | --- | --- |
| 목적 | 비슷한 데이터 그룹화 | 특성 수 줄이기 |
| 예시 | 고객을 3개 그룹으로 분류 | 100개 특성 → 2개로 축소 |

# 강화학습 (Reinforcement Learning)
에이전트가 환경과 상호작용하면서 보상을 최대화하는 행동을 학습하는 방법이다.

정답이 주어지는 것이 아니라 행동에 대한 보상(reward)과 벌칙(penalty)을 통해 학습한다.

- 게임 AI (알파고, 아타리 게임 등)
- 로봇 제어
- 자율 주행

| 구성 요소 | 설명 |
| --- | --- |
| 에이전트(Agent) | 학습하고 행동하는 주체 |
| 환경(Environment) | 에이전트가 상호작용하는 대상 |
| 상태(State) | 현재 환경의 상황 |
| 행동(Action) | 에이전트가 취하는 행동 |
| 보상(Reward) | 행동에 대한 피드백 |

# 머신러닝 워크플로우

1. **데이터 수집** - 문제에 맞는 데이터를 모은다.
2. **데이터 전처리** - 결측치 처리, 스케일링, 인코딩 등
3. **데이터 분리** - 학습 데이터와 테스트 데이터로 나눈다.
4. **모델 선택** - 문제에 적합한 알고리즘을 선택한다.
5. **모델 학습** - 학습 데이터로 모델을 훈련시킨다.
6. **모델 평가** - 테스트 데이터로 성능을 측정한다.
7. **하이퍼파라미터 튜닝** - 성능을 개선하기 위해 파라미터를 조정한다.

| 단계 | 지도학습 | 비지도학습 |
| --- | --- | --- |
| 데이터 | X, y 필요 | X만 필요 |
| 학습 | 정답과 비교하며 학습 | 데이터의 패턴을 스스로 학습 |
| 평가 | 정답과 예측값 비교 | 군집의 품질 등으로 평가 |
