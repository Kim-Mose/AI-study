# 랜덤포레스트 (Random Forest)
과적합을 피하기 위해 여러 트리를 만들어서 다수결로 결정하는 모델

데이터를 랜덤으로 뽑은 후 특성도 랜덤으로 뽑는다.
랜덤으로 뽑은 데이터와 특성으로 트리를 만든 후 분류는 다수결, 회귀는 평균으로 값을 예측한다.

랜덤한 데이터와 특성을 이용하여 여러 트리를 만들기 때문에 과적합 위험에서 벗어날 수 있다.

## Sklearn으로 구현해보기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
```

```python
data = load_iris()

X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= 0.7)

model = RandomForestClassifier()

model.fit(X_train, y_train)

y_pred= model.predict(X_test)

print(accuracy_score(y_test, y_pred))
```

```python
plt.figure(figsize=(15, 8))
plot_tree(model.estimators_[0], filled=True)  # 0번째 트리
plt.show()
```

## RandomForest의 하이퍼파라미터 알아보기

| 파라미터명 | 기본값 | 설명 |
| --- | --- | --- |
| n_estimators | 100 | 트리 개수 |
| max_depth | None | 트리의 최대 깊이 |
| min_samples_split | 2 | 노드를 나누기 위한 최소 데이터 수 |
| min_samples_leaf | 1 | 리프 노드의 최소 데이터 수 |
| max_features | "sqrt" | 분기 시 사용할 특성 수 |
| random_state | None | 랜덤 시드 |
| criterion | Classifier : gini, Regressor : MSE | 평가 지표 |
