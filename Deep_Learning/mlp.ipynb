# 다층 퍼셉트론 (MLP, Multi-Layer Perceptron)
여러 개의 퍼셉트론을 층으로 쌓은 신경망 구조이다.

단층 퍼셉트론이 풀지 못하던 비선형 문제(XOR 등)를 해결할 수 있다.

## MLP의 구조
- **입력층 (Input Layer)** : 데이터를 받는 층
- **은닉층 (Hidden Layer)** : 입력에서 특성을 추출하는 층 (1개 이상)
- **출력층 (Output Layer)** : 최종 결과를 출력하는 층

각 층의 모든 뉴런이 다음 층의 모든 뉴런과 연결되어 있어 **완전연결층(Fully Connected Layer)** 이라고 한다.

## 순전파 (Forward Propagation)
입력 → 은닉층 → 출력층 방향으로 계산이 진행되는 과정

각 층마다 다음 연산을 수행한다.

$$z = Wx + b$$
$$a = f(z)$$

- $W$ : 가중치 행렬
- $b$ : 편향
- $f$ : 활성화 함수
- $a$ : 다음 층의 입력

## PyTorch로 MLP 구현

```python
import torch
import torch.nn as nn

class MLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, output_size)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

model = MLP(input_size=4, hidden_size=16, output_size=3)
print(model)
```

## XOR 게이트 학습
단층 퍼셉트론으로는 풀 수 없는 XOR 문제를 MLP로 해결할 수 있다.

```python
X = torch.tensor([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
y = torch.tensor([[0.], [1.], [1.], [0.]])

class XORModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(2, 4)
        self.fc2 = nn.Linear(4, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x

model = XORModel()
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.05)

for epoch in range(2000):
    pred = model(X)
    loss = criterion(pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

print(model(X).round())
```
