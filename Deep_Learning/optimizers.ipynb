# 옵티마이저 (Optimizer)
역전파로 계산한 기울기를 이용해 가중치를 업데이트하는 알고리즘이다.

기본은 경사하강법이지만, 학습 속도와 안정성을 개선하기 위해 다양한 옵티마이저가 등장했다.

## 주요 옵티마이저

| 옵티마이저 | 특징 |
| --- | --- |
| SGD | 가장 기본, 학습률에 민감 |
| Momentum | 이전 기울기 방향을 고려해 관성 추가 |
| AdaGrad | 자주 업데이트되는 가중치는 작게 학습 |
| RMSProp | AdaGrad의 학습률이 너무 작아지는 문제 개선 |
| Adam | Momentum + RMSProp, 가장 많이 사용 |

## SGD (Stochastic Gradient Descent)
$$w_{t+1} = w_t - \alpha \frac{\partial L}{\partial w}$$

가장 기본적인 옵티마이저로, 단순하지만 학습이 느리거나 불안정할 수 있다.

## Momentum
이전 기울기를 일정 비율로 누적해서 관성처럼 사용한다.

$$v_{t+1} = \beta v_t + \alpha \frac{\partial L}{\partial w}$$
$$w_{t+1} = w_t - v_{t+1}$$

학습이 빠르고 진동이 줄어든다.

## Adam
Momentum과 RMSProp을 결합한 옵티마이저로, 실무에서 가장 많이 사용된다.

- 학습률 자동 조정
- 빠른 수렴
- 대부분의 문제에서 잘 동작

## PyTorch에서의 옵티마이저

```python
import torch
import torch.nn as nn

model = nn.Linear(10, 1)

# SGD
optimizer_sgd = torch.optim.SGD(model.parameters(), lr=0.01)

# SGD + Momentum
optimizer_momentum = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

# Adam
optimizer_adam = torch.optim.Adam(model.parameters(), lr=0.001)

# RMSProp
optimizer_rms = torch.optim.RMSprop(model.parameters(), lr=0.01)
```

## 어떤 옵티마이저를 써야 할까?
- **기본**: Adam (대부분 잘 동작)
- **이미지 분류 (CNN)**: SGD + Momentum (Adam보다 일반화 성능 좋음)
- **빠른 프로토타입**: Adam
