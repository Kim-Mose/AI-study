# 퍼셉트론 (Perceptron)
신경망의 가장 기본 단위로, 여러 입력을 받아 하나의 출력을 만드는 모델이다.

$$y = f(w_1 x_1 + w_2 x_2 + \cdots + w_n x_n + b)$$

- $x_i$ : 입력
- $w_i$ : 가중치
- $b$ : 편향
- $f$ : 활성화 함수

## 퍼셉트론의 구조
1. 입력에 가중치를 곱한다.
2. 모두 더하고 편향을 더한다.
3. 활성화 함수를 통과시킨다.
4. 결과를 출력한다.

선형 회귀나 로지스틱 회귀와 구조가 비슷하다.

## 단층 퍼셉트론의 한계
단층 퍼셉트론은 **선형 분리 가능한 문제**만 풀 수 있다.

AND, OR 게이트는 풀 수 있지만 **XOR 게이트**는 풀 수 없다.

이 문제를 해결하기 위해 **다층 퍼셉트론(MLP)** 이 등장했다.

## PyTorch로 단층 퍼셉트론 구현

```python
import torch
import torch.nn as nn

# 단층 퍼셉트론 정의
class Perceptron(nn.Module):
    def __init__(self, input_size):
        super().__init__()
        self.linear = nn.Linear(input_size, 1)

    def forward(self, x):
        return torch.sigmoid(self.linear(x))

model = Perceptron(input_size=2)
print(model)
```

```python
# AND 게이트 학습 예시
X = torch.tensor([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
y = torch.tensor([[0.], [0.], [0.], [1.]])

model = Perceptron(input_size=2)
criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

for epoch in range(1000):
    pred = model(X)
    loss = criterion(pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

print(model(X).round())
```
