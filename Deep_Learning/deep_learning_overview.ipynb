# 딥러닝 개요 (Deep Learning Overview)
딥러닝은 인공 신경망(Neural Network)을 기반으로 데이터를 학습하는 머신러닝의 한 분야이다.

## 머신러닝 vs 딥러닝

| 구분 | 머신러닝 | 딥러닝 |
| --- | --- | --- |
| 특성 추출 | 사람이 직접 설계 | 모델이 자동으로 학습 |
| 데이터 양 | 적어도 동작 | 많을수록 좋음 |
| 연산량 | 비교적 적음 | 매우 많음 (GPU 필요) |
| 해석 가능성 | 높음 | 낮음 (블랙박스) |
| 예시 | 선형회귀, SVM, RandomForest | CNN, RNN, Transformer |

## 신경망의 구조
신경망은 입력층, 은닉층, 출력층으로 구성된다.

| 층 | 역할 |
| --- | --- |
| 입력층 (Input Layer) | 데이터를 입력받는 층 |
| 은닉층 (Hidden Layer) | 입력을 변환하며 특성을 추출하는 층 |
| 출력층 (Output Layer) | 최종 예측값을 출력하는 층 |

은닉층이 2개 이상이면 **딥러닝(Deep Learning)** 이라고 부른다.

## 딥러닝의 학습 과정
1. **순전파 (Forward Propagation)** - 입력에서 출력까지 계산
2. **손실 계산 (Loss)** - 예측값과 실제값의 차이 측정
3. **역전파 (Backpropagation)** - 손실을 기반으로 가중치 기울기 계산
4. **가중치 업데이트 (Optimizer)** - 경사하강법으로 가중치 조정
5. **반복** - 손실이 충분히 작아질 때까지 1~4 반복

## 주요 딥러닝 모델

| 모델 | 설명 | 사용 분야 |
| --- | --- | --- |
| MLP | 다층 퍼셉트론 (기본 신경망) | 일반적인 분류/회귀 |
| CNN | 합성곱 신경망 | 이미지 처리 |
| RNN | 순환 신경망 | 시퀀스 데이터 |
| LSTM | 장기 기억 신경망 | 긴 시퀀스 데이터 |
| Transformer | 어텐션 기반 모델 | 자연어 처리, 이미지 |

## 딥러닝 프레임워크

| 프레임워크 | 특징 |
| --- | --- |
| PyTorch | 동적 그래프, 연구용으로 인기 |
| TensorFlow | 정적 그래프, 산업 현장에서 많이 사용 |
