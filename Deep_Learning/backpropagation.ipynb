# 역전파 (Backpropagation)
신경망의 출력에서 발생한 오차를 입력 방향으로 거슬러 올라가며 각 가중치의 기울기를 계산하는 알고리즘이다.

이 기울기를 이용해 경사하강법으로 가중치를 업데이트한다.

## 학습 과정
1. **순전파** : 입력 → 출력 방향으로 예측값 계산
2. **손실 계산** : 예측값과 실제값의 차이 측정
3. **역전파** : 출력 → 입력 방향으로 기울기 계산 (연쇄 법칙)
4. **가중치 업데이트** : 옵티마이저로 가중치 조정

## 연쇄 법칙 (Chain Rule)
역전파의 핵심은 **합성 함수의 미분**인 연쇄 법칙이다.

$$\frac{\partial L}{\partial w} = \frac{\partial L}{\partial y} \cdot \frac{\partial y}{\partial z} \cdot \frac{\partial z}{\partial w}$$

출력에서 손실에 대한 기울기를 구하고, 그 기울기를 거슬러 올라가며 각 층의 가중치에 대한 기울기를 계산한다.

## 기울기 소실 문제 (Vanishing Gradient)
층이 깊어질수록 역전파 과정에서 기울기가 점점 작아져 0에 수렴하는 현상이다.

- 시그모이드, 하이퍼볼릭 탄젠트 활성화 함수에서 발생
- ReLU 활성화 함수, 배치 정규화, 잔차 연결(ResNet) 등으로 완화 가능

## PyTorch에서의 역전파
PyTorch는 자동 미분(Autograd) 기능을 제공해 역전파를 자동으로 수행한다.

```python
import torch

# requires_grad=True로 설정하면 기울기를 추적한다.
x = torch.tensor(2.0, requires_grad=True)
y = x ** 2 + 3 * x + 1

# 역전파 실행
y.backward()

# x에 대한 기울기 (dy/dx = 2x + 3 = 7)
print(x.grad)
```

```python
import torch.nn as nn

# 모델 학습 시 역전파 사용 예시
model = nn.Linear(2, 1)
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

X = torch.tensor([[1., 2.], [3., 4.]])
y = torch.tensor([[5.], [11.]])

for epoch in range(100):
    pred = model(X)               # 순전파
    loss = criterion(pred, y)     # 손실 계산

    optimizer.zero_grad()         # 이전 기울기 초기화
    loss.backward()               # 역전파 (기울기 계산)
    optimizer.step()              # 가중치 업데이트

print(model(X))
```
