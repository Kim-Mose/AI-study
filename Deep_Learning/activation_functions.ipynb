# 활성화 함수 (Activation Function)
신경망의 출력을 비선형으로 변환하는 함수이다.

활성화 함수가 없다면 신경망은 아무리 깊게 쌓아도 결국 하나의 선형 함수가 되어 비선형 문제를 풀 수 없다.

## 주요 활성화 함수

| 함수 | 수식 | 특징 |
| --- | --- | --- |
| Sigmoid | $\sigma(x) = \frac{1}{1+e^{-x}}$ | 0~1 출력, 기울기 소실 문제 |
| Tanh | $\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$ | -1~1 출력, 0 중심 |
| ReLU | $f(x) = \max(0, x)$ | 가장 많이 사용, 음수는 0 |
| Leaky ReLU | $f(x) = \max(0.01x, x)$ | ReLU의 음수 부분 보완 |
| Softmax | $\frac{e^{x_i}}{\sum e^{x_j}}$ | 다중 분류의 출력층 |

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)

sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)
relu = np.maximum(0, x)
leaky_relu = np.where(x > 0, x, 0.01 * x)

fig, axes = plt.subplots(1, 4, figsize=(16, 4))

axes[0].plot(x, sigmoid)
axes[0].set_title("Sigmoid")

axes[1].plot(x, tanh)
axes[1].set_title("Tanh")

axes[2].plot(x, relu)
axes[2].set_title("ReLU")

axes[3].plot(x, leaky_relu)
axes[3].set_title("Leaky ReLU")

plt.tight_layout()
plt.show()
```

## 어떤 활성화 함수를 써야 할까?

| 상황 | 활성화 함수 |
| --- | --- |
| 은닉층 | ReLU (기본), Leaky ReLU |
| 이진 분류 출력층 | Sigmoid |
| 다중 분류 출력층 | Softmax |
| 회귀 출력층 | 활성화 함수 없음 (Linear) |

## PyTorch에서의 활성화 함수

```python
import torch
import torch.nn as nn

x = torch.tensor([-2., -1., 0., 1., 2.])

print("Sigmoid:", torch.sigmoid(x))
print("Tanh:", torch.tanh(x))
print("ReLU:", torch.relu(x))
print("Softmax:", torch.softmax(x, dim=0))
```
