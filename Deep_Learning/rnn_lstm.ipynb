# RNN / LSTM
순서가 있는 데이터(시퀀스)를 처리하기 위한 신경망이다.

- 텍스트, 음성, 시계열 데이터에 사용
- 이전 시점의 정보를 기억하면서 다음 시점을 예측

# RNN (Recurrent Neural Network)
이전 시점의 출력을 다음 시점의 입력으로 사용하는 신경망이다.

$$h_t = \tanh(W_x x_t + W_h h_{t-1} + b)$$

- $x_t$ : 현재 시점 입력
- $h_{t-1}$ : 이전 시점의 은닉 상태
- $h_t$ : 현재 시점의 은닉 상태

## RNN의 한계
**장기 의존성 문제 (Long-term Dependency)**
- 시퀀스가 길어질수록 앞쪽 정보가 뒤쪽까지 전달되지 못함
- 기울기 소실 문제로 학습이 어려움

이를 해결하기 위해 **LSTM**이 등장했다.

# LSTM (Long Short-Term Memory)
RNN의 장기 의존성 문제를 해결한 모델이다.

**셀 상태(Cell State)** 와 **3개의 게이트**를 통해 정보를 선택적으로 기억하고 잊는다.

| 게이트 | 역할 |
| --- | --- |
| 망각 게이트 (Forget Gate) | 이전 정보를 얼마나 잊을지 결정 |
| 입력 게이트 (Input Gate) | 새로운 정보를 얼마나 저장할지 결정 |
| 출력 게이트 (Output Gate) | 어떤 정보를 출력할지 결정 |

# GRU (Gated Recurrent Unit)
LSTM을 단순화한 모델로, 게이트를 2개로 줄였다.
- LSTM보다 학습이 빠름
- 성능은 LSTM과 비슷함

## PyTorch로 RNN/LSTM 구현

```python
import torch
import torch.nn as nn

class RNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        self.rnn = nn.RNN(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.rnn(x)
        out = self.fc(out[:, -1, :])  # 마지막 시점의 출력만 사용
        return out

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])
        return out

rnn_model = RNNModel(input_size=10, hidden_size=20, output_size=1)
lstm_model = LSTMModel(input_size=10, hidden_size=20, output_size=1)

print(rnn_model)
print(lstm_model)
```

## 요즘은 Transformer
RNN/LSTM은 순차적으로 처리해야 해서 병렬화가 어려운데,<br>
**Transformer** 모델이 등장하면서 자연어 처리의 주류는 Transformer로 옮겨갔다.

(GPT, BERT 등이 모두 Transformer 기반)
