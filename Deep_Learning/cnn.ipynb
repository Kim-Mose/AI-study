# CNN (Convolutional Neural Network)
이미지 처리에 특화된 신경망으로, 합성곱 연산을 통해 이미지의 특성을 추출한다.

MLP는 이미지를 1차원으로 펼쳐야 해서 공간 정보가 손실되지만, CNN은 2차원 구조를 그대로 유지한다.

## CNN의 주요 구성 요소

| 구성 요소 | 역할 |
| --- | --- |
| 합성곱 층 (Conv) | 필터를 이용해 특성 맵 생성 |
| 풀링 층 (Pooling) | 특성 맵 크기 축소 (Max, Avg) |
| 완전연결 층 (FC) | 최종 분류/회귀 |

## 합성곱 (Convolution)
필터(커널)를 이미지 위로 슬라이딩하며 곱하고 더해서 새로운 특성 맵을 만든다.

- **필터 크기 (Kernel Size)** : 보통 3x3, 5x5
- **스트라이드 (Stride)** : 필터를 이동하는 간격
- **패딩 (Padding)** : 출력 크기 유지를 위해 가장자리에 0을 추가

## 풀링 (Pooling)
특성 맵의 크기를 줄여 계산량을 감소시키고 중요한 특성만 남긴다.

- **Max Pooling** : 영역 내 최댓값 (가장 많이 사용)
- **Average Pooling** : 영역 내 평균값

## CNN의 일반적인 구조
```
입력 → [Conv → ReLU → Pool] × N → Flatten → FC → 출력
```

## PyTorch로 CNN 구현

````python
import torch
import torch.nn as nn

class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        # 입력: (batch, 1, 28, 28) - 흑백 이미지
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.fc1 = nn.Linear(32 * 7 * 7, 128)
        self.fc2 = nn.Linear(128, 10)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))  # (batch, 16, 14, 14)
        x = self.pool(self.relu(self.conv2(x)))  # (batch, 32, 7, 7)
        x = x.view(x.size(0), -1)                # Flatten
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = CNN()
print(model)
````

## 유명한 CNN 모델
