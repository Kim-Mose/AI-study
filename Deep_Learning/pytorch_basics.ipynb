## 텐서 (Tensor)
PyTorch의 기본 자료구조로, NumPy 배열과 비슷하지만 GPU 연산이 가능하다.

````python
import torch

# 텐서 생성
a = torch.tensor([1, 2, 3])
b = torch.zeros(3, 4)        # 0으로 채운 텐서
c = torch.ones(3, 4)         # 1로 채운 텐서
d = torch.randn(3, 4)        # 정규분포 랜덤

print(a)
print(b.shape)
print(c.dtype)
````

````python
# 텐서 연산
x = torch.tensor([[1., 2.], [3., 4.]])
y = torch.tensor([[5., 6.], [7., 8.]])

print(x + y)        # 덧셈
print(x * y)        # 원소별 곱셈
print(x @ y)        # 행렬 곱셈
print(x.T)          # 전치
````

## 자동 미분 (Autograd)
`requires_grad=True`로 설정하면 텐서에 대한 연산이 추적되고, `backward()`로 자동으로 기울기가 계산된다.

````python
x = torch.tensor(2.0, requires_grad=True)
y = x ** 2 + 3 * x + 1
y.backward()

print(x.grad)  # dy/dx = 2x + 3 = 7
````

## 모델 정의
PyTorch에서 모델은 `nn.Module`을 상속받아 정의한다.

````python
import torch.nn as nn

class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(10, 20)
        self.fc2 = nn.Linear(20, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = Model()
print(model)
````

## 학습 루프
PyTorch의 표준 학습 루프 구조

````python
model = Model()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

X = torch.randn(100, 10)
y = torch.randn(100, 1)

for epoch in range(100):
    pred = model(X)               # 1. 순전파
    loss = criterion(pred, y)     # 2. 손실 계산

    optimizer.zero_grad()         # 3. 기울기 초기화
    loss.backward()               # 4. 역전파
    optimizer.step()              # 5. 가중치 업데이트

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
````

## 모델 저장 / 불러오기

````python
# 저장
torch.save(model.state_dict(), "model.pth")

# 불러오기
model = Model()
model.load_state_dict(torch.load("model.pth"))
model.eval()
````
