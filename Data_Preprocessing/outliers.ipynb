# 이상치 처리 (Outliers)
다른 데이터들과 동떨어진 값을 이상치라고 한다.

이상치는 모델 학습에 큰 영향을 주므로 적절히 처리해야 한다.

## 이상치 탐지 방법

| 방법 | 설명 |
| --- | --- |
| IQR (사분위수 범위) | Q1 - 1.5 × IQR ~ Q3 + 1.5 × IQR 벗어나면 이상치 |
| Z-score | 평균에서 표준편차 3배 이상 벗어나면 이상치 |
| 박스플롯 | 시각적으로 이상치 확인 |
| Isolation Forest | ML 기반 이상치 탐지 |

## 1. IQR 방식

```python
import numpy as np
import pandas as pd

data = [10, 12, 14, 15, 16, 18, 20, 22, 24, 100]  # 100이 이상치
s = pd.Series(data)

Q1 = s.quantile(0.25)
Q3 = s.quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = s[(s < lower) | (s > upper)]
print(f"이상치: {outliers.values}")
```

## 2. Z-score 방식

```python
from scipy import stats

data = [10, 12, 14, 15, 16, 18, 20, 22, 24, 100]
s = pd.Series(data)

z_scores = np.abs(stats.zscore(s))
outliers = s[z_scores > 2]
print(f"이상치: {outliers.values}")
```

## 3. 박스플롯으로 시각화

```python
import matplotlib.pyplot as plt

data = [10, 12, 14, 15, 16, 18, 20, 22, 24, 100]

plt.boxplot(data)
plt.show()
```

## 이상치 처리 방법

| 방법 | 설명 |
| --- | --- |
| 제거 | 이상치 행 삭제 |
| 대체 | 평균, 중앙값, 경계값으로 대체 |
| 변환 | 로그 변환 등으로 영향 줄임 |
| 유지 | 의미 있는 이상치라면 유지 |

**주의**: 이상치를 무조건 제거하면 안 된다. 도메인 지식으로 판단해야 한다.<br>
예) 부자의 소득은 평균보다 매우 높지만 이상치가 아니다.
