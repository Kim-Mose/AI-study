# 데이터 분할 (Data Split)
모델의 일반화 성능을 정확히 측정하기 위해 데이터를 학습용과 평가용으로 나눈다.

## 데이터 분할 종류

| 분할 | 역할 | 비율 (예시) |
| --- | --- | --- |
| 학습 데이터 (Train) | 모델 학습 | 70% |
| 검증 데이터 (Validation) | 하이퍼파라미터 튜닝 | 15% |
| 테스트 데이터 (Test) | 최종 성능 평가 | 15% |

데이터가 적으면 7:3, 8:2로 학습/테스트만 나누기도 한다.

## 1. train_test_split

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

data = load_iris()
X, y = data.data, data.target

# 7:3 비율로 학습/테스트 분리
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    train_size=0.7,
    random_state=42  # 결과 재현용
)

print(X_train.shape, X_test.shape)
```

## 2. 학습/검증/테스트 분할
train_test_split을 두 번 호출하면 된다.

```python
# 1차 분할 : 학습+검증 vs 테스트
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.15)

# 2차 분할 : 학습 vs 검증
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.176)
# 0.176 = 0.15 / 0.85 (전체의 15%가 되도록)

print(X_train.shape, X_val.shape, X_test.shape)
```

## 3. Stratified Split (계층적 분할)
분류 문제에서 클래스 비율을 유지하며 분할한다.

특히 클래스가 불균형할 때 중요하다.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,         # 타겟의 비율을 유지
    random_state=42
)
```

## 주의사항

1. **테스트 데이터를 학습에 절대 사용하지 않는다.**<br>
   스케일링, 인코딩 등 전처리도 학습 데이터로만 fit해야 한다.

2. **시계열 데이터는 random하게 나누면 안 된다.**<br>
   과거 데이터로 학습, 미래 데이터로 테스트해야 한다.

3. **random_state를 지정하면 결과가 재현 가능하다.**
