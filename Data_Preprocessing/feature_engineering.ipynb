# 피처 엔지니어링 (Feature Engineering)
기존 데이터를 가공하여 모델 성능을 높이는 새로운 특성을 만드는 작업이다.

데이터 사이언스에서는 "좋은 모델보다 좋은 피처가 더 중요하다"고 할 정도로 중요하다.

## 주요 기법

| 기법 | 설명 |
| --- | --- |
| 파생 변수 생성 | 기존 변수를 조합/변환하여 새 변수 생성 |
| 비닝 (Binning) | 연속형 변수를 구간으로 나눔 |
| 로그 변환 | 분포가 치우친 데이터를 정규분포에 가깝게 |
| 다항 변수 | $x, x^2, x_1 \times x_2$ 등 |
| 날짜 분해 | 날짜를 년, 월, 일, 요일로 분리 |

## 1. 파생 변수 생성

```python
import pandas as pd

df = pd.DataFrame({
    "키": [160, 170, 180, 175],
    "몸무게": [50, 65, 80, 70]
})

# BMI 추가
df["BMI"] = df["몸무게"] / (df["키"] / 100) ** 2
print(df)
```

## 2. 비닝 (Binning)
연속형 변수를 카테고리로 변환

```python
df = pd.DataFrame({"나이": [10, 25, 35, 45, 55, 65, 75]})

# 동일 간격으로 나누기
df["나이대"] = pd.cut(df["나이"], bins=[0, 20, 40, 60, 100], labels=["청소년", "청년", "중년", "노년"])
print(df)
```

## 3. 로그 변환
치우친 분포를 정규분포에 가깝게 만든다.

```python
import numpy as np

df = pd.DataFrame({"소득": [1000, 2000, 5000, 10000, 50000, 100000]})

df["log_소득"] = np.log1p(df["소득"])  # log(1+x)
print(df)
```

## 4. 날짜 분해

```python
df = pd.DataFrame({"날짜": pd.to_datetime(["2024-01-15", "2024-06-20", "2024-12-25"])})

df["년"] = df["날짜"].dt.year
df["월"] = df["날짜"].dt.month
df["일"] = df["날짜"].dt.day
df["요일"] = df["날짜"].dt.dayofweek

print(df)
```

## 5. 다항 변수
기존 변수의 제곱, 곱셈 등으로 새 변수 생성

```python
from sklearn.preprocessing import PolynomialFeatures

X = [[1, 2], [3, 4], [5, 6]]

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

print(X_poly)
# 1, x1, x2, x1^2, x1*x2, x2^2
```

## 피처 엔지니어링의 핵심
**도메인 지식이 가장 중요하다.**

- 어떤 변수가 의미 있는지
- 어떤 조합이 타겟과 관련 있는지
- 데이터를 어떻게 변환해야 더 잘 표현되는지

도메인 지식 없이는 좋은 피처를 만들기 어렵다.
