# 스케일링 (Scaling)
특성들의 값 범위를 비슷하게 맞추는 작업이다.

예를 들어 키(150~190)와 몸무게(40~100)는 단위가 달라 그대로 학습하면 큰 값을 가진 특성에 모델이 편향될 수 있다.

## 스케일링이 필요한 모델 / 불필요한 모델

| 필요 | 불필요 |
| --- | --- |
| KNN | 의사결정나무 |
| SVM | 랜덤포레스트 |
| 신경망 | XGBoost |
| 선형 회귀 (정규화 사용 시) | |

거리/내적 기반 알고리즘과 경사하강법 기반 모델은 스케일링이 필수다.

## 주요 스케일러

| 스케일러 | 수식 | 특징 |
| --- | --- | --- |
| StandardScaler | $\frac{x - \mu}{\sigma}$ | 평균 0, 표준편차 1 |
| MinMaxScaler | $\frac{x - x_{min}}{x_{max} - x_{min}}$ | 0~1 사이로 변환 |
| RobustScaler | $\frac{x - median}{IQR}$ | 이상치에 강함 |
| MaxAbsScaler | $\frac{x}{|x_{max}|}$ | -1~1 사이로 변환 |

## 1. StandardScaler

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

X = np.array([[160, 50], [170, 65], [180, 80], [190, 95]])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled)
print("평균:", X_scaled.mean(axis=0))
print("표준편차:", X_scaled.std(axis=0))
```

## 2. MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

X = np.array([[160, 50], [170, 65], [180, 80], [190, 95]])

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled)  # 모든 값이 0~1 사이
```

## 3. RobustScaler
이상치에 강한 스케일러

```python
from sklearn.preprocessing import RobustScaler

X = np.array([[160, 50], [170, 65], [180, 80], [190, 95], [300, 200]])  # 이상치 포함

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled)
```

## 주의사항
**테스트 데이터에는 fit하지 않는다.**

학습 데이터로 fit한 스케일러를 그대로 테스트 데이터에 transform만 해야 한다.

```python
from sklearn.model_selection import train_test_split

X = np.array([[160, 50], [170, 65], [180, 80], [190, 95], [185, 78]])
y = np.array([0, 0, 1, 1, 1])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # 학습 데이터로만 fit
X_test_scaled = scaler.transform(X_test)        # 테스트 데이터는 transform만
```
