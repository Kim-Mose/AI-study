# 결측치 처리 (Missing Values)
데이터에 비어있는 값(NaN)을 처리하는 방법이다.

결측치가 있으면 모델 학습이 안 되거나 성능이 떨어지므로 반드시 처리해야 한다.

## 결측치 확인

```python
import numpy as np
import pandas as pd

data = {
    "이름": ["철수", "영희", "민수", "지혜"],
    "나이": [25, np.nan, 30, 28],
    "점수": [80, 90, np.nan, 85]
}

df = pd.DataFrame(data)

# 결측치 확인
print(df.isnull())
print(df.isnull().sum())  # 컬럼별 결측치 개수
```

## 결측치 처리 방법

| 방법 | 설명 | 사용 상황 |
| --- | --- | --- |
| 제거 | 결측치 있는 행/열 삭제 | 결측치가 적을 때 |
| 평균/중앙값 대체 | 평균이나 중앙값으로 채움 | 수치형 데이터 |
| 최빈값 대체 | 가장 많은 값으로 채움 | 범주형 데이터 |
| 보간 | 앞뒤 값으로 추정해서 채움 | 시계열 데이터 |
| 모델 예측 | 다른 특성으로 예측해서 채움 | 정교한 처리 필요시 |

## 1. 결측치 제거

```python
# 결측치가 있는 행 제거
df.dropna()

# 결측치가 있는 열 제거
df.dropna(axis=1)

# 모든 값이 결측치인 행만 제거
df.dropna(how="all")
```

## 2. 평균/중앙값 대체

```python
# 평균으로 대체
df["나이"] = df["나이"].fillna(df["나이"].mean())

# 중앙값으로 대체
df["점수"] = df["점수"].fillna(df["점수"].median())

print(df)
```

## 3. 보간 (Interpolation)
시계열 데이터에서 앞뒤 값으로 추정해서 채운다.

```python
s = pd.Series([1, 2, np.nan, np.nan, 5, 6])
print(s.interpolate())  # 1, 2, 3, 4, 5, 6
```

## 4. sklearn의 SimpleImputer

```python
from sklearn.impute import SimpleImputer

# 평균으로 대체
imputer = SimpleImputer(strategy="mean")  # mean, median, most_frequent, constant
X = [[1, 2], [np.nan, 3], [7, 6]]
print(imputer.fit_transform(X))
```
