# 인코딩 (Encoding)
범주형(카테고리) 데이터를 숫자로 변환하는 작업이다.

대부분의 머신러닝 모델은 숫자만 입력으로 받기 때문에 문자열 데이터는 인코딩이 필요하다.

## 주요 인코딩 방법

| 방법 | 설명 | 사용 상황 |
| --- | --- | --- |
| Label Encoding | 카테고리에 정수를 부여 | 순서가 있는 데이터 |
| One-Hot Encoding | 카테고리마다 0/1 컬럼 생성 | 순서가 없는 데이터 |
| Ordinal Encoding | 정해진 순서대로 정수 부여 | 등급, 학년 등 |
| Target Encoding | 타겟의 평균값으로 인코딩 | 카테고리가 많을 때 |

## 1. Label Encoding
각 카테고리에 0, 1, 2, ... 정수를 부여한다.

```python
from sklearn.preprocessing import LabelEncoder

data = ["사과", "바나나", "사과", "포도", "바나나"]

encoder = LabelEncoder()
encoded = encoder.fit_transform(data)

print(encoded)              # [0 1 0 2 1]
print(encoder.classes_)     # ['바나나' '사과' '포도']
```

**주의** : 사과(0), 바나나(1), 포도(2)에는 순서나 크기 의미가 없는데<br>
모델은 0 < 1 < 2로 인식할 수 있어 잘못된 학습이 될 수 있다.

→ 이런 경우 **One-Hot Encoding**을 써야 한다.

## 2. One-Hot Encoding
각 카테고리마다 별도의 컬럼을 만들어 0 또는 1로 표현한다.

```python
import pandas as pd

df = pd.DataFrame({"과일": ["사과", "바나나", "사과", "포도", "바나나"]})

# pandas의 get_dummies
encoded = pd.get_dummies(df["과일"])
print(encoded)
```

```python
from sklearn.preprocessing import OneHotEncoder

data = [["사과"], ["바나나"], ["사과"], ["포도"], ["바나나"]]

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(data)

print(encoded)
print(encoder.categories_)
```

## 3. Ordinal Encoding
순서가 있는 카테고리에 사용한다. (예: 학년, 등급)

```python
from sklearn.preprocessing import OrdinalEncoder

data = [["고"], ["중"], ["저"], ["고"], ["중"]]

encoder = OrdinalEncoder(categories=[["저", "중", "고"]])  # 순서 지정
encoded = encoder.fit_transform(data)

print(encoded)  # 저=0, 중=1, 고=2
```

## 어떤 인코딩을 써야 할까?
- **순서가 있는 데이터** (학년, 등급 등) → Ordinal Encoding
- **순서가 없고 카테고리가 적음** (성별, 색깔 등) → One-Hot Encoding
- **타겟 변수** (분류 라벨) → Label Encoding
- **카테고리가 매우 많음** (도시명 1000개 등) → Target Encoding
